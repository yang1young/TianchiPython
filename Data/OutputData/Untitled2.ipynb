```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime

    
def showCate_level_id(cate_level_ID,number):
    df = pd.read_csv("2_isf2.csv",index_col=0)
    df39 = df[df['cate_level_id']==cate_level_ID]
    df39_date = df39.date.value_counts()
    df39_date.sort_index()
    df39_date = pd.DataFrame(df39_date)
    df39_date.columns = ['num']
    df39_date['date'] = df39_date.index
    days_20141009 = datetime.datetime(2014, 10, 9)
    df39_date['days_20141009'] = df39_date['date'].map(lambda x:(datetime.datetime(x/10000,x/100%100,x%100)-days_20141009).days)
    plt.plot(df39_date.days_20141009,df39_date.num,'ro')
    #plt.show()
    plt.savefig('/Users/zhuohaizhen/Desktop/BigClass/'+str(number)+"-"+str(cate_level_ID)+".jpg")
    plt.show()
    plt.close('all')
```

```python
showCate_level_id(12,1)
showCate_level_id(11,2)
showCate_level_id(8,3)
showCate_level_id(1,4)
showCate_level_id(13,5)
showCate_level_id(4,6)
showCate_level_id(5,7)
showCate_level_id(7,8)
showCate_level_id(10,9)
showCate_level_id(3,10)
showCate_level_id(6,11)
```
